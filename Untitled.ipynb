```python
def down_sample(data, npoint, RAN=True):
    data_down = torch.from_numpy(data)
    data_down = torch.unsqueeze(data_down, 0)
    data_down = data_down.to(torch.float32)
    data_down_idx = utils.farthest_point_sample(data_down, npoint, RAN=True)
    data_down = utils.index_points(data_down, data_down_idx)
    data_down = torch.squeeze(data_down, 0)
    return data_down.numpy()
```

```python
dataset_root = r'H:\irregular construct\PF-Net-Point-Fractal-Network-master\dataset\timber_axis'
filelist1 = os.listdir(dataset_root)
for filename1 in filelist1:
    path1 = os.path.join(dataset_root, filename1)
    filelist2 = os.listdir(path1)
    for filename2 in filelist2:
        path2 = os.path.join(path1, filename2)
        filelist = os.listdir(path2)
        for file in filelist:
            file_path = os.path.join(path2, file)
            data = np.loadtxt(file_path, delimiter=',', dtype=float)
            data_down = down_sample(data, 2048)
            new_path = os.path.join(r'H:\irregular construct\PF-Net-Point-Fractal-Network-master\dataset\downscale_2048')
            if not os.path.exists(os.path.join(new_path, filename1, filename2)):
                os.makedirs(os.path.join(new_path, filename1, filename2))
            np.savetxt(os.path.join(new_path, filename1, filename2, file), data_down, fmt = "%f,%f,%f")
```
